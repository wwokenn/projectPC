Виведіть на екран результати виконання наступних ФУНКЦІЙ КОРИСТУВАЧА:

1. ФУНКЦІЯ користувача збільшує число у 2 рази, а потім додає 4.

2. ФУНКЦІЯ користувача просить користувача ввести слово, додає до нього знак оклику і виводить його...

3. ФУНКЦІЯ користувача просить користувача ввести число, виводить "Дякую!" і повертає це число.

```python
# 1
def double_and_add_four(number): # збільшує вхідне число у 2 рази і додає 4 до результату.
    result = number * 2 + 4
    return result

# 2
def add_exclamation(word): # додає знак оклику до введеного слова.
    word_with_exclamation = word + "!"
    return word_with_exclamation

# 3
def thank_you_and_return_number(): # просить користувача ввести число, виводить "Дякую!" і повертає введене число.
    user_input = input("Введіть число: ")
    number = float(user_input)  # Перетворюємо введену стрічку на число
    print("Дякую!")
    return number

# Виклик функцій і виведення результатів
number1 = 10
result1 = double_and_add_four(number1)
print(f"Результат функції 1 для числа {number1}: {result1}")

word2 = input("Введіть слово: ")
result2 = add_exclamation(word2)
print(f"Результат функції 2: {result2}")

result3 = thank_you_and_return_number()
print(f"Результат функції 3: {result3}")
```

2) Скласти простий калькулятор, де кожній операції відповідає окрема ФУНКЦІЯ користувача (користувачеві пропонується ввести знак, потім перше і друге число. Потім користувач отримує результат, який обчислюється у функції користувача). Якщо користувач введе слово "exit", виконання буде припинено). Використовуйте шаблон:
# while True:
# inputSymbol = ...
# if inputSymbol == "exit":
# break 

```python
def add(x, y): 
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Помилка: ділення на нуль не можливе!"
    return x / y

while True:
    print("Виберіть операцію:")
    print("1. Додавання")
    print("2. Віднімання")
    print("3. Множення")
    print("4. Ділення")
    print("Для виходу введіть 'exit'")
    
    input_symbol = input("Ваш вибір: ")
    
    if input_symbol == "exit":
        print("Ви закрили калькулятор.")
        break
    
    if input_symbol in ('1', '2', '3', '4'):
        num1 = float(input("Введіть перше число: "))
        num2 = float(input("Введіть друге число: "))
        
        if input_symbol == '1':
            result = add(num1, num2)
            print("Результат:", result)
        elif input_symbol == '2':
            result = subtract(num1, num2)
            print("Результат:", result)
        elif input_symbol == '3':
            result = multiply(num1, num2)
            print("Результат:", result)
        elif input_symbol == '4':
            result = divide(num1, num2)
            print("Результат:", result)
    else:
        print("Некоректний вибір операції. Спробуйте ще раз.")
```

5. Інженерний калькулятор:

Давайте перетворимо ваш минулий калькулятор на інженерний. Додайте блоки elif для кожної кнопки та створіть відповідні користувацькі функції. З'ясуйте, які математичні методи вам знадобляться. Завдання виконується за варіантами з 1 по 3. Будь ласка, розподіліть номери варіантів відповідно до списку групи.

```python
import math

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Помилка: ділення на нуль"
    return x / y

def square_root(x):
    if x < 0:
        return "Помилка: корінь з від'ємного числа"
    return math.sqrt(x)

def power(x, y):
    return x ** y

def logarithm(x, base):
    if x <= 0 or base <= 0 or base == 1:
        return "Помилка: некоректний вхід"
    return math.log(x, base)

while True:
    print("Виберіть операцію:")
    print("1. Додавання")
    print("2. Віднімання")
    print("3. Множення")
    print("4. Ділення")
    print("5. Квадратний корінь")
    print("6. Піднесення до ступеня")
    print("7. Логарифм")
    print("Для виходу введіть 'exit'")
    
    input_symbol = input("Ваш вибір: ")
    
    if input_symbol == "exit":
        break
    
    if input_symbol in ('1', '2', '3', '4', '5', '6', '7'):
        if input_symbol in ('1', '2', '3', '4', '6'):
            num1 = float(input("Введіть перше число: "))
            num2 = float(input("Введіть друге число: "))
        
        if input_symbol == '1':
            result = add(num1, num2)
        elif input_symbol == '2':
            result = subtract(num1, num2)
        elif input_symbol == '3':
            result = multiply(num1, num2)
        elif input_symbol == '4':
            result = divide(num1, num2)
        elif input_symbol == '5':
            num1 = float(input("Введіть число: "))
            result = square_root(num1)
        elif input_symbol == '6':
            num1 = float(input("Введіть основне число: "))
            num2 = float(input("Введіть ступінь: "))
            result = power(num1, num2)
        elif input_symbol == '7':
            num1 = float(input("Введіть число: "))
            base = float(input("Введіть основу логарифму: "))
            result = logarithm(num1, base)
            
        print("Результат:", result)
    else:
        print("Некоректний вибір операції. Спробуйте ще раз.")
```
